```python
response_commit = requests.get("https://api.github.com/repos/ironhack-datalabs/madrid-oct-2018/commits")

results_commits = response_commit.json()
results_commits
```

```python
from datetime import *
from pandas.io.json import json_normalize

normalized_commits = json_normalize(results_commits)
# first_day = pd.to_datetime("2019-01-01T00:00:00Z")
# last_day = pd.to_datetime("2019-01-05T00:00:00Z")

first_day = pd.to_datetime("2019-07-08T00:00:00Z")
last_day = pd.to_datetime("2019-07-14T00:00:00Z")

commits_df
commits_df = pd.DataFrame(normalized_commits)

commits_df["commit.committer.date"] = pd.to_datetime(commits_df["commit.committer.date"])
# commits_df.dtypes

commits_df_last_week = commits_df[(commits_df["commit.committer.date"] > first_day) & (commits_df["commit.committer.date"] < last_day)]
commits_df_last_week
```

```python

n_commits = len(commits_df_last_week)

print(f"The number of commits made last week is: {n_commits}")
```

```python
#######Challenge 3########
# https://github.com/ironhack-datalabs/scavenger
#/repos/:owner/:repo/git/trees/:tree_sha?recursive=1
import re
import base64


response = requests.get("https://api.github.com/repos/ironhack-datalabs/scavenger/git/trees/master?recursive=1")
# response = requests.get("https://api.github.com/repos/ironhack-datalabs/scavenger/tree/master?recursive=1")
# response
result_trees = response.json()
# result_trees

result_trees_flattered = json_normalize(result_trees, "tree")
result_trees_flattered["new"] = result_trees_flattered["path"].str.contains("\w+.scavengerhunt", regex = True)
result_trees_flattered_final = result_trees_flattered[result_trees_flattered["new"] == True]
archivos = list(result_trees_flattered_final['path'])


test = []


for file in archivos:
    read_file = requests.get("https://api.github.com/repos/ironhack-datalabs/scavenger/contents/" + file)
    test.append(read_file)
    
resultado = []    
for file in test:
    resultado.append(file.json())
    
resultado_df = pd.DataFrame(resultado)
resultado_df.sort_values(by = "name", inplace = True)
resultado_df
   

```

```python
for index, row in resultado_df.iterrows():
    content = row["content"]
    resultado_df["content"][index] = base64.b64decode(row["content"])    

```

```python
resultado_df
#(?<=b')\w+
# resultado_df["decoded_content"] = resultado_df["content"].str.contains(".*", regex = True)
resultado_df["content"] = resultado_df["content"].astype(str)
for index, row in resultado_df.iterrows():
#     print(type(row["content"]))
#     print(resultado_df["content"][index])
    resultado_df["content"][index] = re.findall(r"(?<=b')\w+", row["content"])

```
